# Autoencoder for the Kolmogorov Flow

In this session, you will need to use what you learned in the previous notebooks to develop a particular neural network called an autoencoder. The dataset comes from the simulation of the Kolmogorov flow:
\begin{equation}
    \nabla \cdot \mathbf{u} = 0
\end{equation}
\begin{equation}
    \frac{\partial \mathbf{u}}{\partial t} + \mathbf{u} \cdot \nabla  \mathbf{u} = -\nabla p + F + \frac{1}{Re} \nabla^2 \mathbf{u}
\end{equation}
where $F=(\sin(2\pi k y/L_y),0)$ where $L_x=L_y=2\pi$ and $k = 4$.

Different sets are provided at different Reynolds number (20, 30 and 40).

## 1. Import librairies

```python
import numpy as np
import tensorflow as tf
from datetime import datetime
import matplotlib.pyplot as plt
import h5py
from tensorflow.keras.layers import Input, Dense, Conv2D, MaxPool2D, UpSampling2D, concatenate, BatchNormalization, Conv2DTranspose, Flatten, Reshape
from numpy.linalg import eigh, svd
```

# 2. Read dataset

```python
hf = h5py.File('02_AutoEncoder_Kolmogorov/Data/ML_WORKSHOP_Kolmogorov_Re20_T1000_DT01.h5','r')
Nx = 24 # Nx=24 for Re=40 or Re=30; Nx=48 for Re=40
Nu = 2
Ntrain = 700 # 700 for Re=20; 7000 for Re=30; 3500 for Re=40
t = np.array(hf.get('t'))
u_all = np.zeros((Nx,Nx,len(t),Nu))
u_all[:,:,:,0] = np.array(hf.get('u_refined'))
if Nu==2:
    u_all[:,:,:,1] = np.array(hf.get('v_refined'))
u_all = np.transpose(u_all,[2,0,1,3])
hf.close()
print(u_all.shape)

# normalize data
u_min = np.amin(u_all[:,:,:,0])
u_max = np.amax(u_all[:,:,:,0])
u_all[:,:,:,0] = (u_all[:,:,:,0] - u_min) / (u_max - u_min)
if Nu==2:
    v_min = np.amin(u_all[:,:,:,1])
    v_max = np.amax(u_all[:,:,:,1])
    u_all[:,:,:,1] = (u_all[:,:,:,1] - v_min) / (v_max - v_min)
```

```python
# visualization of the dataset
%matplotlib inline
fig = plt.figure()
ax = fig.add_subplot(121)
ax.imshow(u_all[0,:,:,0])
ax2 = fig.add_subplot(122)
ax2.imshow(u_all[0,:,:,1])
```

# 4. Define the network
Define the autoencoder and all its layers.
Please not that ultimately you will need to have separately the decoder and encoder part of the autoencoder.

```python
# define the encoder, decoder and autoencoder

#encoder
input_shape = Input(shape = (Nx,Nx,Nu), name='Input')

x = Conv2D(16, (2,2), padding='same', activation = 'linear', name='Conv1')(input_shape)
x = Conv2D(16, (2,2), padding='same', activation = 'linear', name='Conv2')(x)

x = Flatten()(x)
encoded = Dense(50, activation = 'linear', name='Bottleneck')(x)

#---------------------------------------------------------------------------#

#Decoder

x = Dense(Nx*Nx*16, activation = 'linear', name='Decoder1')(encoded)
x = Reshape((Nx, Nx, 16))(x)

x = Conv2DTranspose(16, (2,2), padding='same', activation = 'tanh', name='TConv1')(x)
#x = Conv2DTranspose(16, (2,2), padding='same', activation = 'linear')(x)
decoded = Conv2DTranspose(2, (2,2), padding='same', activation = 'linear', name='TConv2')(x)

autoencoder = tf.keras.Model(input_shape, decoded, name='autoencoder')
autoencoder.summary()
```

```python
lin_en = Flatten()(input_shape)
x = Dense(50, activation = 'linear')(lin_en)

#---------------------------------------------------------------------------#

#Decoder
x = Dense(Nx*Nx*Nu, activation = 'linear')(x)
lin_dec = Reshape((Nx, Nx, Nu))(x)

lin_model = tf.keras.Model(input_shape, lin_dec)
lin_model.summary()
```

# 7. Compute POD modes

```python
dim = u_all.shape
UU = np.reshape(u_train,(Ntrain, dim[1]*dim[2]*dim[3]))
```

```python
# Compute covariance matrix
C = UU.T.dot(UU)/(Ntrain-1)

#solve eigenvalue problem, check LA.eigh
w, W = eigh(C)

#Sort Eigenvalues and vectors
w = np.flip(w)
W = np.fliplr(W)
WT = W.T

#project onto modes for temporal coefficients
a = UU.dot(W)
```

```python
print("relative contribution of eigenvalues")
plt.figure(figsize=(10, 5))
plt.semilogy(w)
#plt.savefig('contrib_ev_20')
plt.figure(figsize=(10, 5))
plt.semilogy(np.cumsum(w)/np.sum(w))
#plt.savefig('cum_ev_20')
```

# 9. Projection of the validation data onto truncated mode and error assessment 

```python


# Mean reconstruction error for different number of retained modes
err = np.zeros((dim[1]*dim[2]*dim[3]))
err_valid = np.zeros((dim[1]*dim[2]*dim[3]))
for i in range(1,(dim[1]*dim[2]*dim[3]) + 1):
    UU_rec = a[:,:i].dot(WT[:i,:])
    err[i - 1] = np.mean((UU - UU_rec)**2)
    UU_rec_val = a_valid[:,:i].dot(WT[:i,:])
    err_valid[i - 1] = np.mean((UU_valid - UU_rec_val)**2)
    if (i%100==0):
        print(i)
```

```python
plt.figure(figsize=(7, 5))
plt.semilogy(err,'b', label= 'train_loss, Re=20')
plt.semilogy(err_valid,'r', label = 'val_loss, Re=20')
plt.title("Loss History")
plt.xlabel("N Modes")
plt.ylabel("Loss")
plt.legend()
#plt.savefig('loss_pod_20.png')
```
